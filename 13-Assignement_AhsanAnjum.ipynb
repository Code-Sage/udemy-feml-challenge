## House Prices dataset

```python
from math import sqrt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import scipy.stats as stats

# for the model
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, r2_score

# for removing outliers in numerical variables
from feature_engine.outlier_removers import Winsorizer

# for feature engineering
from sklearn.preprocessing import StandardScaler, RobustScaler
from feature_engine import missing_data_imputers as mdi
from feature_engine import discretisers as dsc
from feature_engine import categorical_encoders as ce
```

```python
# ignoring some pesky clutter of warnings

import warnings
warnings.filterwarnings('ignore', category=UserWarning)
warnings.filterwarnings('ignore', category=FutureWarning)
```

### Load Datasets

```python
# load dataset

data = pd.read_csv('../houseprice.csv', index_col='Id')
```

```python
data_vars = list(data.nunique().sort_values().index)
```

```python
# make lists of variable types

categorical = [var for var in data.columns if data[var].dtype == 'O']

year_vars = [var for var in data.columns if 'Yr' in var or 'Year' in var]

discrete = [
    var for var in data.columns if data[var].dtype != 'O'
    and len(data[var].unique()) < 20 and var not in year_vars
]

numerical = [
    var for var in data.columns if data[var].dtype != 'O'
    if var not in discrete and var not in ['Id', 'SalePrice']
    and var not in year_vars
]

print('There are {} continuous variables'.format(len(numerical)))
print('There are {} discrete variables'.format(len(discrete)))
print('There are {} temporal variables'.format(len(year_vars)))
print('There are {} categorical variables'.format(len(categorical)))
```

## Adding Boolean Features

```python
data['HasWoodDeck'] = np.where(data.WoodDeckSF > 0, 1, 0)
data['HasMasonryVanir'] = np.where(data.MasVnrArea > 0, 1, 0)
data['HasOpenPorch'] = np.where(data.OpenPorchSF > 0, 1, 0)
data['Has3SsnPorch'] = np.where(data['3SsnPorch'] > 0, 1, 0)
data['HasScreenPorch'] = np.where(data.ScreenPorch > 0, 1, 0)
data['HasEnclosedPorch'] = np.where(data.EnclosedPorch > 0, 1, 0)
data['HasAiryPorch'] = data.HasOpenPorch | data.HasScreenPorch
data['HasShelteredPorch'] = data.HasEnclosedPorch | data.Has3SsnPorch
```

### Separate train and test set

```python
# IMPORTANT: keep the random_state to zero for reproducibility
# Let's separate into train and test set

X_train, X_test, y_train, y_test = train_test_split(data.drop(
    ['SalePrice'], axis=1),
    data['SalePrice'],
    test_size=0.1,
    random_state=0)
```

```python
# calculate elapsed time

def elapsed_years(df, var):
    # capture difference between year variable and
    # year the house was sold
    
    df[var] = df['YrSold'] - df[var]
    return df

for var in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    X_train = elapsed_years(X_train, var)
    X_test = elapsed_years(X_test, var)
```

```python
# drop YrSold
X_train.drop('YrSold', axis=1, inplace=True)
X_test.drop('YrSold', axis=1, inplace=True)
```

```python
# capture the column names for use later in the notebook
final_columns = X_train.columns
```

## Feature Engineering Pipeline

```python
# I will treat discrete variables as if they were categorical
# to treat discrete as categorical using Feature-engine
# we need to re-cast them as object

X_train[discrete] = X_train[discrete].astype('O')
X_test[discrete] = X_test[discrete].astype('O')
```

```python
# categorical variables having # of missing values > 0
categorical_w_missing = ['Alley', 'MasVnrType', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1',
                         'BsmtFinType2', 'Electrical', 'FireplaceQu', 'GarageType', 'GarageFinish',
                         'GarageQual', 'GarageCond', 'PoolQC', 'Fence', 'MiscFeature']
```

```python
house_pipe = Pipeline([
    # missing data imputation - section 4
    ('missing_ind',
     mdi.AddMissingIndicator(variables=['LotFrontage', 'MasVnrArea', 'GarageYrBlt'] + categorical_w_missing)),
    
    # median imputation for numerical variables
    ('imputer_num',
     mdi.MeanMedianImputer(
         imputation_method='median',
         variables=['LotFrontage', 'MasVnrArea', 'GarageYrBlt'])),
    
    # removing outliers from numerical variables
    ('winsorize', Winsorizer(distribution='skewed', tail='both', fold=3, variables=numerical + year_vars[:-1])),
    
    # replacing missing values in categorical variables by frequent categories
    ('imputer_cat', mdi.CategoricalVariableImputer(imputation_method='frequent', variables=categorical)),

    
    # categorical encoding - section 6
    ('rare_label_enc',
     ce.RareLabelCategoricalEncoder(tol=0.15,
                                    n_categories=6,
                                    variables=categorical + discrete)),

    ('categorical_enc',
     ce.MeanCategoricalEncoder(variables=categorical + discrete)),
     

    # discretisation + encoding - section 8
    ('discretisation', dsc.DecisionTreeDiscretiser(cv=3, variables=numerical)),
    

    # feature Scaling - section 10
    ('scaler', RobustScaler()),
    
    
    # regression
    ('lasso', Lasso(random_state=0))
])
```

```python
# let's fit the pipeline
house_pipe.fit(X_train, y_train)

# let's get the predictions
X_train_preds = house_pipe.predict(X_train)
X_test_preds = house_pipe.predict(X_test)
```

## latest / last-run

```python
# check model performance:

print('train mse: {}'.format(mean_squared_error(y_train, X_train_preds)))
print('train rmse: {}'.format(sqrt(mean_squared_error(y_train, X_train_preds))))
print('train r2: {}'.format(r2_score(y_train, X_train_preds)))
print()
print('test mse: {}'.format(mean_squared_error(y_test, X_test_preds)))
print('test rmse: {}'.format(sqrt(mean_squared_error(y_test, X_test_preds))))
print('test r2: {}'.format(r2_score(y_test, X_test_preds)))
```

```python
# plot predictions vs real value

plt.scatter(y_test,X_test_preds)
plt.xlabel('True Price')
plt.ylabel('Predicted Price')
```

```python
# let's explore the importance of the features
# the importance is given by the absolute value of the coefficient
# assigned by the Lasso

importance = pd.Series(np.abs(house_pipe.named_steps['lasso'].coef_))
importance.index = list(final_columns)+[i + '_na' for i in ['LotFrontage', 'MasVnrArea',  'GarageYrBlt'] + categorical_w_missing]
importance.sort_values(inplace=True, ascending=False)
importance.plot.bar(figsize=(18,6))
```
